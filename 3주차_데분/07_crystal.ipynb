## col 추가

```python
df['정식호칭'] = df['약칭'] + ' ' + df['이름']
df
```

```python
df['집권방식'] = '민선'
df
```

```python
df.loc[df['집권기간']>5, '집권방식'] = '독재'
df.loc[(df['집권기간']<3), '집권방식'] = '임시'
df.loc[(df['집권기간']<5) & (df['집권방식']=='민선'), '집권방식'] = '탄핵'
df.loc[(df['집권기간']==0) & (df['집권방식']=='임시'), '집권방식'] = '낙선'
df
```

## col 삭제

```python
df = df.drop(columns=['정식호칭', '집권방식'])
df
```

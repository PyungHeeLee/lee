```python
data['이름']
```

## df 객체 생성

```python
df = pd.DataFrame(data)
df
```

```python
print(df['이름'], df[['이름', '키']])
```

## df 객체 생성 (인덱스 지정)

```python
df = pd.DataFrame(data, index = ['1~3대', '4대', '5~9대', '10대', '11~12대', '13대', '14대', '15대', '낙선1', '16대', '17대', '18대', '19대', '20대', '낙선2'])
```

```python
df
```

```python
df
```

# 3. 인덱스

```python
df = pd.DataFrame(data, index = ['1~3대', '4대', '5~9대', '10대', '11~12대', '13대', '14대', '15대', '낙선', '16대', '17대', '18대', '19대', '20대', '낙선'])
df
```

## 인덱스 이름 설정

```python
df.index.name = '번호'
df
```

## 인덱스 초기화

```python
df.reset_index(drop=True)
```

```python
df.reset_index(drop=True, inplace=True)
# 이래야 반영
```

## 인덱스 설정

```python
df.set_index('이름', inplace=True)
df
# 바로 반영
```

## 인덱스 기준으로 데이터 정렬

```python
df.sort_index()
# 오름차순
```

```python
df.sort_index(ascending=False)
# 오름차순이 아니다 -> 내림차순
```

```python
import pandas as pd
```

```python
data = {
    '이름' : ['이승만', '윤보선', '박정희', '최규하', '전두환', '노태우', '김영삼', '김대중', '이회창', '노무현', '이명박', '박근혜', '문재인', '윤석열', '이재명'],
    '키': [172, 169, 162, 182, 172, 180, 172, 170, 168, 165, 175, 163, 173, 178, 175],
    '출생연도': [1870, 1890, 1910, 1920, 1930, 1930, 1920, 1920, 1930, 1940, 1940, 1950, 1950, 1960, 1960],
    '출신지': ['황해', '충남', '경북', '강원', '경남', '경북', '경남', '전남', '황해', '경남', '경북', '경북', '경남', '서울', '경북'],
    '집권기간' : [9, 0.5, 18, 0.5, 7, 5, 5, 5, 0, 5, 5, 4.5, 5, 5, 0],
    '약칭': ['Lee', 'Yoon', 'President Part', 'Choi', 'Jeon', 'Noe', 'Sam', 'Kim', 'Chang', 'Noe', 'Lee', 'Park', 'Moon', 'Seok', 'Lee']
}
```

```python
df = pd.DataFrame(data)
df.index.name = '번호'
df
```

## 저장 to csv

```python
df.to_csv('presidents1.csv', encoding='utf-8-sig')
# py 파일 있는 곳에 바로 만들어짐
```

```python
df.to_csv('presidents2.csv', encoding='utf-8-sig', index=False)
```

## 저장 to txt

```python
df.to_csv('presidents3.txt', sep='\t')
# tab으로 구분된 txt 파일
```

## 저장 to 엑셀

```python
df.to_excel('presidents4.xlsx')
```

## 열기 from csv

```python
df = pd.read_csv('presidents1.csv')
df
```

```python
df = pd.read_csv('presidents1.csv', skiprows=2)
df
# 무시할 rows 수 정의
```

```python
df = pd.read_csv('presidents1.csv', skiprows=[2,3])
df
# 1,3 번째 행 무시
```

```python
df = pd.read_csv('presidents1.csv', nrows=3)
df
# 3개만 가져옴
```

```python
df = pd.read_csv('presidents1.csv', skiprows=1, nrows=2)
df
# 처음 한개 무시, 그 이후 두개 표시
```

## 열기 from txt

```python
df = pd.read_csv('presidents3.txt', sep='\t')
df
```

```python
df = pd.read_csv('presidents3.txt', sep='\t', index_col='번호')
df
```

```python
df = pd.read_csv('presidents3.txt', sep='\t')
df.set_index('번호', inplace=True)
df
```

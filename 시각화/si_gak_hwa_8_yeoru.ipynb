```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 글자가 깨지는 현상을 해결
```

```python
df = pd.read_excel('../pandas/suneung.xlsx', index_col='번호')
df
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,5))
fig.suptitle('각 과목 성적')

axs[0].bar(df.index, df['국어'], label='백분위')
axs[0].set_title('국어')
axs[0].legend()
axs[0].set(xlabel='번호', ylabel='백분위')
axs[0].set_facecolor('lightyellow')
axs[0].grid(linestyle='--', linewidth=0.3)

axs[1].plot(df.index, df['수학'], label='백분위')
axs[1].set_title('수학')
axs[1].legend()
axs[1].set(xlabel='번호', ylabel='백분위')

axs[2].barh(df.index, df['탐구'], label='백분위')
axs[2].set_title('탐구')
axs[2].legend()
axs[2].set(xlabel='백분위', ylabel='번호')

plt.show()
```

## loc를 통한 데이터 선택
인덱스 이름을 이용 -> 원하는 row에서 col 선택

```python
df.loc[2]
# 번호 2번 출력
# 인덱스의 자료형을 무조건 따를 것 (지금은 int)
```

```python
df.loc[2, '집권기간']
# 박정희 집권기간
```

```python
df.loc[[3,4], ['출신지', '키']]
```

```python
df.loc[0:2, '키':'출신지']
```

## iloc을 통한 데이터 선택
위치를 이용

```python
df.iloc[0]
# 첫번째 데이터
# 인덱스의 자료형 무시하고 int index 사용
```

```python
df.iloc[7:]
```

```python
df.iloc[5, 0]
# 6번째 대통령의 이름
```

```python
df.iloc[[4,9], 3:5]
```

## 조건에 따른 데이터 선택

```python
df['출신지'] == '경북'
```

```python
filt = df['출신지'] == '경북'
df[filt]
# 경북 출신 대통령
```

```python
df[-filt]
```

```python
df[df['키']<168]
```

```python
df.loc[df['키']<168, '집권기간']
```

```python
df.loc[(df['키']>168) & (df['출신지']=='경남')]
```

```python
df.loc[(df['키']>179) | (df['키']<165)]
```

## str 함수

```python
filt = df['이름'].str.startswith('이')
df[filt]
```

```python
filt = df['이름'].str.contains('박', na=True)
df[filt]
# NaN에 대해서는 True 취급
```

```python
local = ['서울', '경남']
filt = df['출신지'].isin(local)
df[filt]
```

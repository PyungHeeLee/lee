## 데이터 채우기 fillna

```python
import pandas as pd
import numpy as np
df = pd.read_excel('presidents4.xlsx', index_col='번호')
print(df,type(df))
```

```python
df.iloc[1,3] = np.nan
df.iloc[3,3] = np.nan
df.iloc[7,3] = np.nan
df.iloc[2,2] = np.nan
df.iloc[4,2] = np.nan
df
# np 이용해서 결측처리
```

```python
df.fillna('없음')
# 결측치 처리
#inplace 처리 해줘야 반영됨
```

```python
df['출신지'] = df['출신지'].fillna('모름')
df
#출신지 결측값만 모름 처리
```

```python
df = pd.read_excel('presidents4.xlsx', index_col='번호')
df
# 원상복구
```

## 결측값 삭제

```python
df.iloc[1,3] = np.nan
df.iloc[3,3] = np.nan
df.iloc[7,3] = np.nan
df.iloc[2,2] = np.nan
df.iloc[4,2] = np.nan
df.iloc[7,2] = np.nan
df.iloc[1,4] = np.nan
df
```

```python
df.dropna()
# 결측있는 행 그냥 삭제
```

```python
df.dropna(axis='index', how='any')
# 한 행에서, 하나라도 결측이 존재하면 그 행 삭제
```

```python
df.dropna(axis='columns', how='all')
# 한 열에서, 모든 행이 결측이어야 그 열 삭제 (현재 경우에선 삭제 X)
```

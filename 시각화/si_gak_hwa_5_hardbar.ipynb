```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 글자가 깨지는 현상을 해결
```

```python
df = pd.read_excel('../pandas/suneung.xlsx', index_col='번호')
df
```

# 누적 bar

```python
plt.bar(df.index, df['국어'], color='r')
plt.bar(df.index, df['수학'], color='b')
# plt.bar(df['이름'], df['키']) -> 이름 표시
```

```python
plt.bar(df['이름'], df['국어'], label='국')
plt.bar(df['이름'], df['수학'], bottom=df['국어'], label='수')
plt.bar(df['이름'], df['탐구'], bottom=df['국어']+df['수학'], label='탐')
plt.xticks(rotation=60)
plt.legend(ncol=3)
```

```python
print(df.shape[0], df.shape[1]) # 각각 행 수, 열 수
```

```python
n = df.shape[0] # n=행 수
n
```

```python
index = np.arange(n)
index
```

```python
w=0.2
plt.bar(index-w, df['국어']) # 두께가 w보다 큼 -> 겹쳐 보임
plt.bar(index, df['수학'])
plt.bar(index+w, df['탐구'])
```

```python
plt.figure(figsize=(10,5))
plt.title('각자 백분위')

w = 0.2
plt.bar(index-w, df['국어'], width=w, label='국')
plt.bar(index, df['수학'], width=w, label='수')
plt.bar(index+w, df['탐구'], width=w, label='탐') #두께만큼 빗겨서 생성 -> 압 겹침
plt.legend(ncol=3, loc='center')
plt.xticks(index, df['이름'], rotation=30)
plt.show()
```

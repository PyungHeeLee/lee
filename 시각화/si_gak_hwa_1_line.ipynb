```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 글자가 깨지는 현상을 해결
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용 가능한 폰트
[f.name for f in fm.fontManager.ttflist]
```

```python
x=[1,2,3]
y=[5,8,9]

plt.plot(x,y, label='범례입니다', linewidth=4, ls='--', color='#ff0000', alpha=0.3)
plt.legend(loc='lower right') # label의 위치: upper left 등
# plt.legend(loc=[0.6,0.3]) -> 0~1로도 설정 가능
# plt.legend(ncol=3) -> 범례의 열 수 정의

plt.title('선')
# plt.title('꺾은선 그래프', fontdict={'family':'HYGungSo-Bold', 'size':20}) -> 개별 폰트

plt.xlabel('x축', color='red', loc='left') # left center right
plt.ylabel('y축', color='#00aa00', loc='bottom') # top center bottom
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.show()
```

```python
plt.plot(x, y, marker='o', ms=15, mec='red', mfc='yellow') # marker: v x X
```

```python
plt.plot(x, y, 'gx-.') # 포맷: color marker linstyle
```

```python
plt.figure(figsize=(2,1))
plt.plot(x, y)
```

```python
plt.figure(figsize=(2,1), dpi=150) # dots per inch
plt.plot(x, y)
```
